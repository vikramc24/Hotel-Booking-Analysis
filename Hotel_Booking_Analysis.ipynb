### Cancellation Rate

Here we find out the cancelation rate of hotel bookings

```python
# find out the total number of hotel bookings

total_booking_numbers = len(hotel_bookings_copy_df)
total_booking_numbers
```

```python
# Now we find out the number of booking cancelled

booking_cancelled = hotel_bookings_copy_df[hotel_bookings_copy_df['is_canceled'] == 1].shape[0]

booking_cancelled
```

```python
# Calculating the rate of cancellation

cancellation_rate = (booking_cancelled/total_booking_numbers)*100

cancellation_rate
```

***the cancellation rate of hotel bookings  is 37.04%***

```python
# Visual presentaion of cancellation rate

labels = ['Canceled', 'Not Canceled']
sizes = [booking_cancelled, total_booking_numbers-booking_cancelled]
colors = ['pink', 'yellow']
explode = (0.1,0)

fig, ax = plt.subplots(figsize=(5,5))

ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=130)

plt.title("Rate of Hotel Booking Cancellation")

plt.show()
```

### Hotel Booking Cancellation Year Wise

```python
cancellation_yearly = hotel_bookings_copy_df.groupby('arrival_date_year').agg(bookings_total= pd.NamedAgg(column = 'is_canceled',aggfunc ='count'),
    booking_canceled = pd.NamedAgg(column = 'is_canceled', aggfunc ='sum')).reset_index()

cancellation_yearly
```

```python
# Calculating the yearly cancellation rate

cancellation_yearly['rate_of_cancellation']= (cancellation_yearly['booking_canceled']/cancellation_yearly['bookings_total'])*100
cancellation_yearly['rate_of_cancellation']
```

```python
# Visual representaion of yearly hotel booking cancellation

plt.figure(figsize=(9,7))
sns.barplot(data= cancellation_yearly, x ='arrival_date_year', y = 'rate_of_cancellation', palette ='viridis')
plt.title("Cancellation Rate Yearwise")
plt.xlabel('year')
plt.ylabel('Rate of Cancellation(In Percentage)')
plt.show()
```

# Revenue Analysis

To conduct a thorough revenue analysis of the hotel booking dataset, we will focus on several key financial matrics and analyse their trends over time. Specifically we will examine:

1- Total Revenue

2- ADR (Average Daily Rate)

3- RevPAR (Revenue per available room

4- Occupancy Rate

```python
# converting necessary columns to appropriate type which we can analyze

hotel_bookings_copy_df['reservation_status_date']=pd.to_datetime(hotel_bookings_copy_df['reservation_status_date'])

## Extract the year

hotel_bookings_copy_df['year']=hotel_bookings_copy_df['reservation_status_date'].dt.year
```

```python
# to simplify we assume the constant number of rooms

# Assume each hotels has 100 rooms

total_rooms = 100
```

```python
# calculate total_revenue, ADR, Occupancy Rate and RevPAR year wise

financial_matrics = hotel_bookings_copy_df.groupby('year').agg({'adr' : 'mean', 'reservation_status_date':'size'}).rename(columns= {'reservation_status_date': 'total_bookings'})

financial_matrics
```

```python
# Calculate total rooms available

financial_matrics['total_rooms']= total_rooms*365 #(365 days in a year


financial_matrics['total_rooms']
```

```python
# Calculating the total revenue of the hotels

financial_matrics['total_revenue']= hotel_bookings_copy_df.groupby('year')['adr'].sum()

financial_matrics['total_revenue']
```

```python
# calculate occupancy rate

financial_matrics['occupancy_rate']= (financial_matrics['total_bookings']/financial_matrics['total_rooms'])*100
financial_matrics['occupancy_rate']
```

```python
# calculating RevPAR (Revenue Per Available Room)

financial_matrics['revPAR'] = financial_matrics['adr'] * (financial_matrics['occupancy_rate']/100)
financial_matrics['revPAR']
```

```python
# Visulisation of total revevue per year

plt.figure(figsize =(10,3))
plt.plot(financial_matrics. index, financial_matrics['total_revenue'], marker = 'o', linestyle = '-', color = 'tab:orange')
plt.title('Total Revenue Over Years')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```

```python
# Visualization of ADR and occupancy rate

fig, ax1= plt.subplots(figsize= (10,3))
color = 'tab:green'

ax1.set_xlabel('year')
ax1.set_ylabel('ADR',color = color)
ax1.plot(financial_matrics.index,financial_matrics['adr'],color=color, label= 'ADR')

ax2= ax1.twinx()
color= 'tab:blue'

ax2.set_ylabel('Occupancy Rate (%)', color = color)

ax2.plot(financial_matrics.index, financial_matrics['occupancy_rate'],color= color, label= 'Occupancy Rate')

ax2.tick_params(axis ='y', labelcolor = color)

fig.tight_layout()
plt.title('ADR and Occupancy Rate Over Years')
plt.show()
```

```python
# visalization of RevPAR (Revenue Per Available Room)

plt.figure(figsize=(10,3))

plt.plot(financial_matrics.index,financial_matrics['revPAR'],marker ='o', linestyle= '-', color = 'tab:red')
plt.title('RevPAR over Years')
plt.xlabel('Year')
plt.ylabel('RevPAR (Revenue Per Available Room)')
plt.grid(True)
plt.show()
```

Based on the analysis we can make several conclusions about the revenue of the hotels which also shows the performance of the hotels over the years

A consistent increase indicates healthy business growth while any dips might indicate external factors affecting the business. An increasing ADR (Average Daily Rate) suggest higher rates charged per room which could be due to higher demand or improved services. Higher occupancy rate typically signifies strong demand and effective marketing strategies. RevPAR (Revenue Per Available Room) is comprehensive metric combining both ADR and Occupancy rates. It provides insight into the overall revenue generating efficiency in hotel.

The revenue analysis provides valuable insights into the financial performance of the hotels. By tracking these matrices over time hotel management can identify trends, make informed strategic decisions and optimize pricing and marketing strategies to enhance revenue and profitability.
